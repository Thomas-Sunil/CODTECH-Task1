```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt
housing = fetch_california_housing()


data = housing.data
target = housing.target

print(data.shape)
print(target.shape)
```

```python
df = pd.DataFrame(data, columns=housing.feature_names)
df['MedHouseVal'] = target
print(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
corr_matrix = df.corr()
print(corr_matrix)
```

```python
import numpy as np
def find_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers
outliers = find_outliers_iqr(df['MedHouseVal'])
print(df[outliers])
```

```python
import seaborn as sns
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
```

```python
plt.figure(figsize=(14, 10))
for i, column in enumerate(df.columns):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9, 6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['MedHouseVal'], cmap='coolwarm', alpha=0.6)
plt.colorbar(label='MedHouseVal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Housing Prices')
plt.grid(True)
plt.show()
```
